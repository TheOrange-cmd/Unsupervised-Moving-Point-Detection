```python
import open3d as o3d
from notebook_helpers import setup_for_notebook, process_and_get_pcd

# --- Configuration ---
CONFIG_FILE = 'best_tuning.yaml'

# Initialize everything (detector, data actor, etc.)
config_accessor, detector, data_actor, all_sweep_tokens = setup_for_notebook(CONFIG_FILE)
```

```python
import k3d
import numpy as np

# --- CHOOSE THE FRAME TO DEBUG ---
# For example, let's look at a frame after the initialization phase
frame_index_to_debug = 30 

# This function will run the detector on all prior frames to build its state,
# then return the point cloud for your target frame.
pcd_to_view = process_and_get_pcd(
    frame_index_to_debug,
    all_sweep_tokens,
    data_actor,
    detector,
    config_accessor
)

# Visualize the result in a new window
if pcd_to_view:
    print(f"Displaying point cloud for frame {frame_index_to_debug} using K3D...")
        
    points = np.asarray(pcd_to_view.points).astype(np.float32)
    colors = np.asarray(pcd_to_view.colors).astype(np.float32) # Cast to float32

    plot = k3d.plot(name='Point Cloud Viewer')
    # This should now work without the TraitError
    plot += k3d.points(positions=points, colors=colors, point_size=0.1) 
    plot.display()
```
